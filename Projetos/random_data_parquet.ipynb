```python
import pandas as pd
import numpy as np
import random

# List of names
names = ['Alice', 'Fernanda', 'Joel', 'Davi', 'Welington', 'Maria', 'João', 'Ana', 'Carlos', 'Beatriz', 'Lucas', 'Julia', 'Rafael', 'Camila', 'Gabriel', 'Luisa', 'Daniel', 'Bruna', 'Felipe', 'Carolina']

# Generate a unique list of numbers
numbers = np.random.choice(range(996261000, 996269999), size=len(names), replace=False)

# List of messages
messages = ['Oi', 'Bom dia', 'Ola', 'Como vai?', 'Bom dia', 'Tudo bem?', 'Como você está?', 'Boa tarde', 'Boa noite', 'Até mais', 'Foi bom te ver', 'Fale com você depois', 'Estou ansioso para nosso encontro', 'Vejo você em breve', 'Tenha um ótimo dia']

# Random data
data = {
    'name': random.choices(names, k=200),
    'number': random.choices(numbers, k=200),
    'message': random.choices(messages, k=200),
    'date': pd.date_range(start='12-12-2023', periods=200).astype(np.int64) // 10**9,  # Convert to Unix timestamp
    'type': ['message'] * 200
}

# Create a DataFrame
df = pd.DataFrame(data)
df = df.sample(frac=1).reset_index(drop=True)  # Shuffle the data
df['date'] = pd.to_datetime(df['date'], unit='s')


print(df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Define o estilo do seaborn
sns.set(style="whitegrid")

# Gráfico de barras para a coluna 'name'
plt.figure(figsize=(10,6))
sns.countplot(x='name', data=df)
plt.title('Frequência de Cada Nome')
plt.xlabel('Nome')
plt.ylabel('Frequência')
plt.xticks(rotation=90)
plt.show()

# Gráfico de barras para a coluna 'message'
plt.figure(figsize=(10,6))
sns.countplot(y='message', data=df)
plt.title('Frequência de Cada Mensagem')
plt.xlabel('Mensagem')
plt.ylabel('Frequência')
plt.show()
```
